# **REGRESI LINNEAR**

```python
from sklearn.datasets import make_regression
from matplotlib import pyplot
X_test, y_test = make_regression(n_samples=30, n_features=1, noise=5)
pyplot.scatter(X_test,y_test)
pyplot.show()
```

```python
import pandas as pd
df = pd.DataFrame(X_test, y_test)
df
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_test, y_test)
reg.score(X_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(np.array([[1]]))
```

```python
pyplot.scatter(X_test, y_test)
Y_plot = reg.coef_*X_test+reg.intercept_

pyplot.plot(X_test, Y_plot, color='r')

pyplot.show()
```
